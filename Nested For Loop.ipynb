# Nested For Loop

```python
for a in range(5):
    print("Inner loop begin")
    for char in "Pakistan":
        print(a,char)
```

```python
country = ["Pakistan","Turkey","Iran","Oman"]
for a in country:
    print(country)
    for number in range (5):
        print(number)
```

## Table Example

```python
tables = int(input("Enter Table number"))
for table in range(2,tables+1):
    for number in range(1,11):
        print(f"{table} * {number} = {table * number}")
```
